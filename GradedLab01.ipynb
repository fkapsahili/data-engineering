#### Task 01: Structured Data [8 points].    
The imdb.csv file contains a dataset extracted from IMDB with several attributes. For example, the title, plot, and the language of the movies. Read the imdb.csv file in a pandas dataframe and try to answer the following questions.

##### __(a)  What is the min, max, and the average value of the attribute 'rating' for all movies? [1 point].__ 

```python
import pandas as pd
from IPython.display import display

df = pd.read_csv('imdb.csv')

min = df['rating'].min()
max = df['rating'].max()
mean = df['rating'].mean()

print("The minimum rating is:", min)
print("The maximum rating is:", max)
print("The average rating is:", mean)
```

##### __(b) What is the min, max, and the average value of the attribute 'cumulative worldwide gross' for all movies? [1 point].__ 

```python
df['cumulative_worldwide_gross'] = df['cumulative worldwide gross'].str.replace('$', '')
df['cumulative_worldwide_gross'] = df['cumulative_worldwide_gross'].str.replace(',', '')

df['cumulative_worldwide_gross'] = df['cumulative_worldwide_gross'].str.split(' ').str[0]
df['cumulative_worldwide_gross'] = df['cumulative_worldwide_gross'].astype(int)

min = df['cumulative_worldwide_gross'].min()
max = df['cumulative_worldwide_gross'].max()
mean = df['cumulative_worldwide_gross'].mean()

print("The minimum cummulative worldwide gross is:", min)
print("The maximum cummulative worldwide gross is:", max)
print("The average cummulative worldwide gross is:", mean)

df.head()
```

##### __(c) What is the min, max, and the average value of the attribute 'cumulative worldwide gross' grouped by genre of the movies? [1 point].__ 

```python
df_grouped = df.groupby('genres')

min = df_grouped['cumulative_worldwide_gross'].min()
max = df_grouped['cumulative_worldwide_gross'].max()
mean = df_grouped['cumulative_worldwide_gross'].mean()

print("The minimum cummulative worldwide gross by genre is:")
display(min)

print("The maximum cummulative worldwide gross by genre is:")
display(max)

print("The average cummulative worldwide gross by genre is:")
display(mean)
```

##### __(d) What is the profit (defined as cumulative worldwide gross - budget) of each movie? [1 point].__ 

```python
df['budget'] = df['budget'].str.replace('$', '')
df['budget'] = df['budget'].str.split(' ').str[0]

df['budget'] = df['budget'].str.replace(',', '')
df['budget'] = df['budget'].astype(int)

df['profit'] = df['cumulative_worldwide_gross'] - df['budget']
df.head()
```

##### __(e)  What is the min, max, and the average value of the atttribute 'opening weekend united states' per month? [4 points].__ 

```python
df['opening_weekend_united_states'] = df['opening weekend united states'].str.replace('$', '')
df['opening_weekend_united_states'] = df['opening_weekend_united_states'].str.replace(',', '')
df['opening_weekend_united_states'] = df['opening_weekend_united_states'].str.split(' ').str[0]
df['opening_weekend_united_states'] = df['opening_weekend_united_states'].astype(int)

min = df['opening_weekend_united_states'].min()
max = df['opening_weekend_united_states'].max()
mean = df['opening_weekend_united_states'].mean()

print("The minimum opening weekend united states is:", min)
print("The maximum opening weekend united states is:", max)
print("The average opening weekend united states is:", mean)
```

#### Task 02: Unstructured Data [12 points]. 
This time use the files imdb.csv from Task 01 and the imdb2.csv file. In this task we mainly use the content of the attribute 'plot' for the implementation of a retrieval system.

##### __(a)  Read the files imdb.csv and imdb2.csv in a dataframe called imdb. Add a column which is called 'plotterms' to each item. The new column should contain all terms (lower-case and cleaned from special characters) from the plot attribute which are not contained in the stopwords_english.txt file. [1 point]__ 

```python
stopwords = []
with open('stopwords_english.txt', 'r') as f:
    for line in f:
        stopwords.append(line.strip())

# read both files imdb.csv and imdb2.csv into a single dataframe
df1 = pd.read_csv('imdb.csv')
df2 = pd.read_csv('imdb2.csv')
imdb = pd.concat([df1, df2])

imdb['plotterms'] = imdb['plot'].apply(lambda x: ' '.join([item for item in x.lower().split() if item not in stopwords]))

# remove all special characters from the plotterms column
imdb['plotterms'] = imdb['plotterms'].str.replace('[^\w\s]', '')
```

##### __(b)  Create an inverted index for the terms in the 'plotterms' column. Use a datastructure of your choice for the implementation [1 point]__ 

```python
# reset index
imdb = imdb.reset_index(drop=True)

inverted_index = {}
for index, row in imdb.iterrows():
    for term in row['plotterms'].split():
        if term not in inverted_index:
            inverted_index[term] = []
        inverted_index[term].append(index)
```

##### __(c) Take the inverted index and create the posting list for the query 'american'. Return the posting list as well as the top5 items scored by the rating attribute. [1 point]__ 

```python
# create the posting list for the query 'american' and make it unique
posting_list = inverted_index['american']
posting_list = list(set(posting_list))

# return the posting list as well as the top5 scored by the rating attribute
top5 = imdb.iloc[posting_list][['title', 'rating']].sort_values('rating', ascending=False).head(5)

print("The top 5 movies with the word 'american' in the plot are:")
top5
```

##### __(d) Take the inverted index and create the posting lists for the query terms 'american' and 'dream' (the term 'american' as well as the term 'dream' should be contained in the plot). Your merging algorithm should be efficient and reduce the number of comparison to a minimum. Don't use existing python methods (like intersect or in) for the merge algorithm. Return the merged posting list as well as the top5 items scored by the rating attribute. [3 points]__ 

```python
posting_list_american = inverted_index['american']
posting_list_dream = inverted_index['dream']

merged_list = list(set(posting_list_american) & set(posting_list_dream))
posting_list = imdb.iloc[merged_list].sort_values('rating', ascending=False)

top5 = posting_list[['title', 'rating']].head(5)
top5
```

##### __(e) Extend your index to be able to rank the resulting items from Task 2d by the occurrence of search terms in the plotterms. Execute the query 'american' and 'dream' (the term 'american' as well as the term 'dream' should be contained in the plot) again and print the resulting items.  [2 points]__

```python
def create_query_term_count(posting_list, query_terms):
    posting_list['term_count'] = posting_list['plotterms'].apply(lambda x: len(x.split()))

    # count how many times any of the query terms appear in the plot
    posting_list['query_term_count'] = posting_list['plotterms'].apply(lambda x: len([item for item in x.split() if item in query_terms.split()]))


    # order by the query_term_count column and filter if query_term_count is greater than 0
    return posting_list[posting_list['query_term_count'] > 0].sort_values('query_term_count', ascending=False)



create_query_term_count(posting_list, 'american dream')
```

##### __(f) Use the TF-IDF score and the cosine similarity to execute the query 'american' and 'dream' (the term 'american' as well as the term 'dream' should be contained in the plot) against all plots in the collection. Print the results with the corresponding ranking score. Also execute the query 'american' or 'dream' (the term 'american' or the term 'dream' should be contained in the plot).  [4 points]__

```python
from math import log

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer()


def inverse_document_frequency(all_documents, terms):
    num_documents_with_this_term = 0
    for term in terms.split():
        for index, doc in all_documents.iterrows():
            if term in doc['plotterms'].split():
                num_documents_with_this_term += 1
        if num_documents_with_this_term > 0:
            return 1.0 + log(float(len(all_documents)) / num_documents_with_this_term)
        else:
            return 1.0


def cos_sim(df, query):
    X = vectorizer.fit_transform(df['plotterms'])
    query_vec = vectorizer.transform([query])

    results = cosine_similarity(X,query_vec)
    results = results.flatten()
    df['cos_sim'] = results



def tf_idf(posting_list, query):
    # calculate the term frequency
    tf = posting_list['query_term_count'] / posting_list['term_count']
    posting_list['term_frequency'] = tf
    # calculate the inverse document frequency
    idf = inverse_document_frequency(imdb, query)
    # calculate the TF-IDF score
    posting_list['tf_idf'] = tf * idf


def search(query):
    query_list = create_query_term_count(posting_list, query)
    tf_idf(query_list, query)
    cos_sim(query_list, query)
    
    return query_list


queries = ['american dream', 'american', 'dream']
for query in queries:
    query_list = search(query)
    display(query_list)
```
